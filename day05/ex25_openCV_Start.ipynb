````python
cap = cv2.VideoCapture('./sample_360p.mp4')

while True:
    ret, frame = cap.read() # FPS의 이미지를 하나씩 읽음
    if not ret:# 더이상 읽을 이미지가 없으면 종료
        break

    cv2.imshow('video', frame) #이미지를 계속 변경

    if cv2.waitKey(1) == ord('q'): # 키보드 Q입력 하면 종료
        break

cap.release()
cv2.destroyAllWindows()
````

````python
# 실시간 웹캠

cap = cv2.VideoCapture(0) # 0은 웹캠 디바이스 번호

while True:
    ret, frame = cap.read()
    if not ret:
        break

    cv2.imshow('video', frame)

    if cv2.waitKey(10) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
````

````python
# 이미지 로드 후 변경
import cv2
cat = cv2.imread('./cat01.jpg',flags = cv2.IMREAD_COLOR_BGR)

# opencv함수로 변경
gray = cv2.cvtColor(cat,cv2.COLOR_BGR2GRAY) # color를 회색톤으로 변경
cv2.imshow('cat', cat)
cv2.imshow('origin',cat)
cv2.imshow('gray',gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
# 채널 분리
import cv2
import numpy as np
(b, g, r) = cv2.split(cat)

# 채널(B,G,R)를 단일 채널로 분리
# np.zeros() : 배열사이즈를 직접 지정해서 0으로 배열초기화
# np.zeros_like(배열) : 지정된 배열사이즈를 0으로 초기화

blue = cv2.merge((b, np.zeros_like(g), np.zeros_like(r)))
green = cv2.merge((np.zeros_like(b), g, np.zeros_like(r)))
red = cv2.merge((np.zeros_like(b), np.zeros_like(g), r))

cv2.imshow('origin', cat)
cv2.imshow('blue', blue)
cv2.imshow('green', green)
cv2.imshow('red', red)

cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
import cv2
cat = cv2.imread('./cat01.jpg',flags = cv2.IMREAD_COLOR_BGR)
small = cv2.resize(cat,None,fx = 0.5, fy = 0.5)
cv2.imshow('origin' ,cat)
cv2.imshow('1/2 size',small)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
#이미지 로드 후 회전
import cv2
cat = cv2.imread('./cat01.jpg',flags = cv2.IMREAD_COLOR_BGR)
rot180 = cv2.rotate(cat,cv2.ROTATE_180)
cv2.imshow('origin' ,cat)
cv2.imshow('rot180',rot180)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
#이미지 로드 후 원 그리기
import cv2
cat = cv2.imread('./cat01.jpg',flags = cv2.IMREAD_COLOR_BGR)
cv2.circle(cat, (416,210), 100, (0,0,255), thickness=3, lineType=cv2.LINE_AA)
cv2.imshow('origin' ,cat)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

#### OpenCV 이미지 처리
- 주요 기능
    - 이미지 읽기, 저장,출력
    - 색상 변환
    - 크기 변경
    - 회전, 이동
    - 블러
    - 이진화
    - 선명도 조절
    - 노이즈 제거
    - 대비/밝기 조절
    - 엣지 검출
    - 윤곽선 찾기
    - 퍼시펙티브 변환 - 원근감 이미지를 평탄화

````python
# 이미지 로드
cat = cv2.imread(filename='./cat01.jpg', flags=cv2.IMREAD_COLOR)

# 텍스트
cv2.putText(cat, 'Kitty', (300, 70),
            cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 255), 4, cv2.LINE_AA)

cv2.imshow('origin', cat)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
# 이미지 로드
cat = cv2.imread(filename='./cat01.jpg', flags=cv2.IMREAD_COLOR)

# ROI 선택
roi = cat[81:81+214, 295:295+240]

cv2.imshow('origin', cat)
cv2.imshow('roi', roi)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
# 이미지 로드
cat = cv2.imread(filename='./cat01.jpg', flags=cv2.IMREAD_COLOR)

## 엣지(이미지 경계)검출
# 그레이스케일 변경
gray = cv2.cvtColor(cat,cv2.COLOR_BGR2GRAY)
canny = cv2.Canny(gray,threshold1 = 100, threshold2 = 200)
cv2.imshow('gray', gray)
cv2.imshow('canny', canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
````
