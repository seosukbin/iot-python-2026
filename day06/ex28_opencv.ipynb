#### 스냅샷 저장

- 동영상 순간 저장 기능

```python
cap = cv2.VideoCapture(0)  # 첫번째 웹캠(0)

prev_time = 0
show_infro = False
while True:
    ret, img = cap.read()
    if not ret:
        break

    curr_time = time.time()

    fps = 1 / (curr_time - prev_time) if prev_time != 0 else 0
    prev_time = curr_time

    fps_text = f'FPS: {int(fps)}'

    cv2.putText(img, fps_text, (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    cv2.imshow('webcam', img)
    key = cv2.waitKey(1)
    if key == ord('q'):
        break
    elif key == ord('s'):
        filename = time.strftime('snapshot_%Y%m%d_%H%M%S.jpg')
        cv2.imwrite(filename, img)
        print(f'{filename} 스냅샷 저장')
cap.release()
cv2.destroyAllWindows()
```

#### 해상도 조절
- 웹캠이 지원하는 해상도로 변경 가능
- 예시: 640x480, 1280x720, 1920x1080 등

```python
import cv2
import time

cap = cv2.VideoCapture(0)  # 첫번째 웹캠(0)

if not cap.isOpened():
    raise RuntimeError('웹캠 오픈 실패!')

# 해상도 설정
WIDTH, HEIGHT = 1280, 720
cap.set(cv2.CAP_PROP_FRAME_WIDTH, WIDTH)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, HEIGHT)

# 실제 적용된 해상도
real_w = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
real_h = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)

prev_time = 0
show_info = True  # 정보 표시 여부

while True:
    ret, img = cap.read()
    if not ret:
        break

    curr_time = time.time()

    fps = 1 / (curr_time - prev_time) if prev_time != 0 else 0
    prev_time = curr_time

    # 정보 표시
    if show_info:
        fps_text = f'FPS: {int(fps)}'
        res_text = f'{int(real_w)}x{int(real_h)}'

        cv2.putText(img, fps_text, (10, 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

        cv2.putText(img, res_text, (10, 70),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    # 화면 출력
    cv2.imshow('webcam', img)

    # 키 입력
    key = cv2.waitKey(1)

    if key == ord('q'):
        break

    elif key == ord('s'):  # 스냅샷 저장
        filename = time.strftime('snapshot_%Y%m%d_%H%M%S.jpg')
        cv2.imwrite(filename, img)
        print(f'{filename} 스냅샷 저장')

    elif key == 9:  # Tab 키 → 정보 on/off
        show_info = not show_info
        print(f'표시여부 {show_info}')

cap.release()
cv2.destroyAllWindows()
```

```python
# Haar Cascade 파일 로드
import cv2
import numpy as np
import time
face_fascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml') # 얼굴 데이터 따로
smile_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_smile.xml') # 미소 데이터 따로


cap = cv2.VideoCapture(0)  # 첫번째 웹캠(0)

if not cap.isOpened():
    raise RuntimeError('웹캠 오픈 실패!')

# 해상도 설정
WIDTH, HEIGHT = 640, 480
cap.set(cv2.CAP_PROP_FRAME_WIDTH, WIDTH)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, HEIGHT)

# 실제 적용된 해상도
real_w = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
real_h = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)

prev_time = 0
show_info = True  # 정보 표시 여부

while True:
    ret, img = cap.read()
    if not ret:
        break

    curr_time = time.time()

    fps = 1 / (curr_time - prev_time) if prev_time != 0 else 0
    prev_time = curr_time

    # 정보 표시
    if show_info:
        fps_text = f'FPS: {int(fps)}'
        res_text = f'{int(real_w)}x{int(real_h)}'

        cv2.putText(img, fps_text, (10, 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

        cv2.putText(img, res_text, (10, 70),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    faces = face_fascade.detectMultiScale(img,1.1,5)

    # 검출된 물체를 표시
    for (x, y, w, h) in faces: # x,y 좌측 상단좌표, w,h는 너비와 높이
        cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 255), 2)

        face_roi = img[y:y+h, x:x+w]

        # 얼굴 내부에서 미소 검출
        small = cv2.resize(face_roi,  (w//15, h//15), interpolation=cv2.INTER_AREA )
        mosaic = cv2.resize(small, (w, h), interpolation=cv2.INTER_NEAREST)
        img[y:y+h, x:x+w] = mosaic

    # 화면 출력
    cv2.imshow('webcam', img)

    # 키 입력
    key = cv2.waitKey(1)

    if key == ord('q'):
        break

    elif key == ord('s'):  # 스냅샷 저장
        filename = time.strftime('snapshot_%Y%m%d_%H%M%S.jpg')
        cv2.imwrite(filename, img)
        print(f'{filename} 스냅샷 저장')

    elif key == 9:  # Tab 키 → 정보 on/off
        show_info = not show_info
        print(f'표시여부 {show_info}')

cap.release()
cv2.destroyAllWindows()
```

# QR 코드 리더

```python
# QR 코드 검출기 객체 생성
qr_detector = cv2.QRCodeDetector()

cap= cv2.VideoCapture(0)
while True:
    ret, img = cap.read()
    if not ret:
        break

    # QR 코드 검출
    data, points, _ = qr_detector.detectAndDecode(img)

    if points is not None and data:
        points = points.astype(int)
        
        for i in range(len(points[0])):
            pt1 = tuple(points[0][i])
            pt2 = tuple(points[0][(i + 1) % len(points[0])])
            cv2.line(img, pt1, pt2, (0, 255, 0), 2)
            
            cv2.putText(img, data, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)

        cv2.imshow('QR Code Scanner', img)

        key = cv2.waitKey(0) 
        if key == ord('q'):
            break           

    cv2.imshow('QR Code Scanner', img)
    if cv2.waitKey(1) == ord('q'):
        break


cap.release()
cv2.destroyAllWindows()
```

# QR리더기 검출

```python
import cv2
import pyzbar.pyzbar as pyzbar  # !pip install pyzbar

# QR코드 검출기 객체 생성
qr_reader = cv2.QRCodeDetector()

# 웹캠 오픈
cap = cv2.VideoCapture(0)

while True:
    ret, img = cap.read()
    if not ret:
        break

    # 회색톤으로 변환
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # pyzbar로 바코드/QR코드 디코딩
    decoded = pyzbar.decode(gray)
    barcode_data = ''

    for d in decoded:
        x, y, w, h = d.rect

        barcode_data = d.data.decode('utf-8')
        barcode_type = d.type

        # 바운딩 박스
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

        # 텍스트 표시
        text = f'{barcode_data} ({barcode_type})'
        cv2.putText(img, text, (x, y),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    cv2.imshow('QR Code Reader', img)

    key = cv2.waitKey(1)
    if key == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import webbrowser

qr_detector = cv2.QRCodeDetector()
cap = cv2.VideoCapture(0)

is_opened = False
last_data = ''
frame_count = 0
detect_interval = 10  # ⭐ while 밖에 선언

data = ''
points = None

while True:
    ret, img = cap.read()
    if not ret:
        break

    frame_count += 1

    # ⭐ 10프레임마다만 QR 검사
    if frame_count % detect_interval == 0:
        data, points, _ = qr_detector.detectAndDecode(img)

    if points is not None and data:
        points = points.astype(int)

        for i in range(len(points[0])):
            pt1 = tuple(points[0][i])
            pt2 = tuple(points[0][(i + 1) % len(points[0])])
            cv2.line(img, pt1, pt2, (0, 255, 0), 2)

        cv2.putText(img, data, (10, 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)

        # URL 열기
        if data.startswith('http://') or data.startswith('https://'):
            if not is_opened or data != last_data:
                webbrowser.open(data)
                is_opened = True
                last_data = data
    else:
        is_opened = False
        last_data = ''

    cv2.imshow('QR Code Scanner', img)

    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
