````python
# matplotlib 라이브러리 로드

import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [10,20,15,30,25]
plt.plot(x,y)
plt.show()
````

## 제목,레이블 표시

````python
# 한글은 기본적으로 깨진다. 안깨지게 하기 위해서는 
plt.plot(x,y)
plt.title('점추 추이')
plt.xlabel('시간')
plt.ylabel('score')
plt.show()
````

- 한글 설정

````python
# 리소스 인수, 폰트 관리자 , 리소스 포함
from matplotlib import rcParams,font_manager,rc
font_path = 'C:/Windows/Fonts/malgun.ttf'
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font',family=font)
rcParams['axes.unicode_minus'] = False # 이걸 넣어 줘야 한다.
````

````python
plt.plot(x,y)
plt.title('점추 추이')
plt.xlabel('시간')
plt.ylabel('score')
plt.show()
````

- 포인트 모양 변경

````python
plt.plot(x,y, marker = 'o') # 마커표시 할떄 사용하는것
plt.title('점추 추이')
plt.xlabel('시간')
plt.ylabel('score')
plt.show()
````

#### 라인,색상,포인트
- x,y 좌표 뒤에 문자열로 작성
    - 색상: b,g,r,c,m,y,k,w
    - 마커: 위와 동일
    - 라인 스타일: -(직선), --(dashed line), -.(dashed dot line), :(dotted line)

````python
plt.plot(x,y,'ro:') # 마커표시 할떄 사용하는것
plt.title('점추 추이')
plt.xlabel('시간')
plt.ylabel('score')
plt.show()
````

````python
#### 3D 그래프

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 리소스 인수, 폰트 관리자 , 리소스 포함
from matplotlib import rcParams,font_manager,rc
font_path = 'C:/Windows/Fonts/malgun.ttf'
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font',family=font)
rcParams['axes.unicode_minus'] = False # 이걸 넣어 줘야 한다.
````

````python
x = [1,2,3,4]
y= [10,20,15,30]
z = [5,10,8,12]

fig = plt.figure() # 새로운 차트 선언해서 변수 할당
ax = fig.add_subplot(111,projection = '3d')

ax.scatter(x,y,z)

ax.set_xlabel('X축')
ax.set_xlabel('Y축')
ax.set_xlabel('Z축')
````
