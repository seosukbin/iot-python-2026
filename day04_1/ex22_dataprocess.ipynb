#### 결측치(중요)
- 실제 data 수집 과정에서 누락 되거나 측정 되지 않는 빈 값
- 데이터 분석에서는 결측치를 처리 하는일이 가장 중요

````python
df['수학_1'] = df['수학'] - 5
df
````

````python
df.sort_values('수학_1')
````

````python
df.sort_values('수학_1',ascending = False)
````

````python
df1 = df.dropna() # 결측치가 있는 데이터행을 아예 없앤다.
df1
````

````python
df2 = df.fillna(45) #결측치가 있는 데이터에 값은 넣는다.
df2
````

````python
# 그룹화(GroupBy)
df.groupby('수학').count()
````

````python
df[['국어','영어','수학']].fillna(45).groupby('수학').mean()
df
````

````python
data = {
'name': ['유고','요미','철수','예성'],
'score': [80,95,60,70]
}

df = pd.DataFrame(data)
df
````

````python
df.to_csv('./scores.csv',encoding = 'utf-8')
````

````python
df[df['score'] >= 70]
````

````python
df['score'].mean()
````

````python
#합격 여부
#리스트 컴프리헨션으로 새 컬럼 생성
['PASS'if s>=70 else 'FAIL' for s in df['score']]
````
