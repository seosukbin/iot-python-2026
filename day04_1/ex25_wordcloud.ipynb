```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = '''python is my favorite programming language, data analysis pyton ai ai ai 
          deep learning, machine learning ai contemporary ai  skill'''

wc = WordCloud().generate(text)
```

```python
plt.imshow(wc)
plt.axis('off')
plt.show()
```

```python
text = '데이터 분석 파이썬 인공지능 인공지능 인공지능 인공지능 데이터 파이썬 인공지능 C++ Java '

wc = WordCloud(font_path = 'C:/Windows/Fonts/malgun.ttf',
background_color = 'azure',colormap = 'cool').generate(text)

plt.imshow(wc)
plt.axis('off')
plt.show
```
